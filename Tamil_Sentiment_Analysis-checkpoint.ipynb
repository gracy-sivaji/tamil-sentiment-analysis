```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded!")
```

```python
df = pd.read_excel("Tamil_Sentiment_Dataset.xlsx")
print("Total records:", len(df))
df.head(10)
```

```python
print("Sentiment Count:")
print(df['Sentiment'].value_counts())
print("\nCategory Count:")
print(df['Category'].value_counts())
```

```python
plt.figure(figsize=(8,5))
colors = ['#70AD47','#FF0000','#FFC000']
df['Sentiment'].value_counts().plot(
    kind='bar',
    color=colors,
    edgecolor='black'
)
plt.title('Tamil Sentiment Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('sentiment_distribution.png')
plt.show()
print("Chart 1 saved!")
```

```python
plt.figure(figsize=(10,5))
df['Category'].value_counts().plot(
    kind='bar',
    color='#1F4E79',
    edgecolor='black'
)
plt.title('Category Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('category_distribution.png')
plt.show()
print("Chart 2 saved!")
```

```python
plt.figure(figsize=(10,6))
pivot = pd.crosstab(df['Category'], df['Sentiment'])
sns.heatmap(
    pivot,
    annot=True,
    fmt='d',
    cmap='YlOrRd',
    linewidths=0.5
)
plt.title('Sentiment by Category Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('sentiment_heatmap.png')
plt.show()
print("Chart 3 saved!")
```

```python
summary = df.groupby(['Category','Sentiment']).size().reset_index(name='Count')
summary.to_excel('sentiment_summary_report.xlsx', index=False)
print("✅ Analysis Complete!")
print("✅ Report saved!")
print("\nProject Summary:")
print(f"Total Sentences: {len(df)}")
print(f"Total Categories: {df['Category'].nunique()}")
print(f"Sentiments: {df['Sentiment'].unique()}")
```
